```python
# generate a list of lats and lons and then add thoes to a tuple
# import numpy
import numpy as np
```

```python
# set the range of lats and longs for United States
latRange = (24, 50)
lngRange = (-125, -66.9)
```

```python
# use np.random.uniform to generate a series of values in range
lats = np.random.uniform(latRange[0], latRange[1], size=50)
lons = np.random.uniform(lngRange[0], lngRange[1], size=50)

#zip the lists of lats and lons into tuple
latLongs = zip(lats, lons)
```

```python
# use nearest city to identify a city near each set of coordinates
cityNames = []

for coords in latLongs:
    # try to get a city name
    cityName = citipy.nearest_city(coords[0], coords[1]).city_name

    # check to see if the city is unique, if so, add to the list
    if cityName not in cityNames:
        cityNames.append(cityName)

for c in cityNames:
    print(c)
```

```python
# build the dataframe of city names and current temperatures, latitudes and longitudes

temperatures = []
lats = []
longs = []
cities = []

queryURL = f"{url}appid={api_key}&units={units}&q="

for city in cityNames:
    try:
        time.sleep(1)  # Pause the program for 1 second
        
        # Create the new query string
        cityQuery = f"{queryURL}{city}"
        
        # Call to requests and jsonify the response
        response = requests.get(cityQuery).json()
        
        # Filter based on latitude, longitude, and country
        if response['coord']['lat'] > 50 or response['coord']['lat'] < 24:
            continue
        if response['coord']['lon'] < -125 or response['coord']['lon'] > -66.9:
            continue
        if response['sys']['country'] != 'US':
            continue

        # Extract the temperature and add to the temperatures list
        temperatures.append(response['main']['temp'])
        
        # Extract the latitude and add to the latitudes list
        lats.append(response['coord']['lat'])
        
        # Extract the longitude and add to the longitudes list
        longs.append(response['coord']['lon'])
        
        # Extract the city name from OpenWeather
        cities.append(response['name'])
    
    except Exception as e:
        # Just in case the temperature or other data is not reported
        print(f"Error fetching data for {city}: {e}")
        pass

# Build the dataframe
cityDataDF = pd.DataFrame(
    {
        "City Name": cities,
        "Fahrenheit Temperature": temperatures,
        "Latitude": lats,
        "Longitude": longs,
    }
)


cityDataDF
```
